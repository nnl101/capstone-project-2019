```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import geopandas as gpd

%matplotlib inline
import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)

# To increase the number of rows pandas will print before truncation
pd.set_option('display.max_rows', 500)

# To increase the number of columns pandas will print before truncation
pd.set_option('display.max_columns', 100)

# To increase the width of each column pandas will print before truncating.
pd.set_option('display.max_colwidth',100)
```

```python
states = gpd.read_file('./data/shapefiles/tl_2017_us_state.shp')
states.rename(columns={'NAME':'State Name'}, inplace=True)
df_states = states[['State Name', 'REGION','DIVISION']]
df_states
df = pd.read_csv('data/VSRR_Provisional_Drug_Overdose_Death_Counts.csv')
df = pd.merge(df, df_states, on='State Name')
```

```python
df['Year'] = df['Year'].astype(str)
df['Month'] = df['Month'].astype(str)
df['Date'] = df['Year'] +'-'+ df['Month']
df['Date'] = [datetime.strptime(x, '%Y-%B') for x in df['Date'] ]
```

```python
df2 = pd.DataFrame(df.groupby(["REGION","Date"])["Predicted Value"].sum().reset_index())
df2
```
