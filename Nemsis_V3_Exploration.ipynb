```
uniq = pd.DataFrame()

for col in stats.columns:
  if stats.loc['unique'][col] <= 15:
    # print(pd.unique(v3_data[col]))
    uniq = uniq.append(pd.Series(data=pd.unique(v3_data[col]),name=col))

uniq
```
