```
sum_by_icd = v3_data.groupby(['Primary Impression ICD Name', 'Primary Impression ICD 10'])['Count of Events'].sum()
opioids_events_df = sum_by_icd.reset_index(level='Primary Impression ICD 10', name='Count of Events')

opioids_events_df
```

```
from math import pi
from bokeh.palettes import Category20c
from bokeh.plotting import figure
from bokeh.transform import cumsum

# data = pd.Series(opioids_events_df['Count of Events']).reset_index(
#     name='value').rename(
#         columns={'Primary Impression ICD Name': 'names'})

data = pd.DataFrame(opioids_events_df).reset_index()

data.rename(columns={'Primary Impression ICD Name': 'names', 'Count of Events': 'value', 'Primary Impression ICD 10' : 'codes'}, inplace=True)

# Filter out Primary Impressions for 'Suicide attempt'
data = data[data['names'] != 'Suicide attempt']
data.sort_values('value', axis='index', ascending=False, inplace=True, na_position='last')
data = data.head(20)

data['angle'] = data['value']/data['value'].sum() * 2*pi
data['color'] = Category20c[20]

p = figure(plot_height=800, plot_width=1300, 
           title="Top 20 Primary Impressions for Opioids", 
          #  toolbar_location=None,
           tools="hover", 
           tooltips="@names (@codes): @value", 
           x_range=(-0.5, 1.0))

p.wedge(x=0, y=1, radius=0.4,
        start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        line_color="white", fill_color='color', legend="names", source=data)

p.axis.axis_label=None
p.axis.visible=False
p.grid.grid_line_color = None

show(p)
```

### Plotting the counts of the top 20 Opioid related primary impressions

```
from bokeh.io import show, output_file
from bokeh.models import ColumnDataSource
from bokeh.palettes import Category20c
from bokeh.plotting import figure
from bokeh.transform import factor_cmap

# output_file("bar_colormapped.html")
source = ColumnDataSource(data)

p = figure(x_range=data.names, plot_height=800, plot_width=800, 
           title="Top 20 Primary Impressions for Opioids",
           tooltips="@names (@codes): @value")
p.vbar(x='names', top='value', width=0.9, source=source,
       line_color='white', 
       fill_color=factor_cmap('names', palette=Category20c[20], factors=data.names))

# p.xgrid.grid_line_color = None
p.xaxis.major_label_orientation = pi/3


show(p)
```
