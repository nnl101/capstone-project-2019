### Pie chart of the counts of the top 20 Opioid related primary impressions

```
from math import pi
from bokeh.palettes import Category20c
from bokeh.plotting import figure
from bokeh.transform import cumsum

data = pd.DataFrame(opioids_events_df).reset_index()
data['names'] = data['Primary Impression ICD Name'].astype(str) + " (" + data['Primary Impression ICD 10'].astype(str) + ")"
data.rename(columns={'Count of Events': 'value'}, inplace=True)

# print(data)

# Filter out Primary Impressions for 'Suicide attempt'
data = data[data['names'] != 'Suicide attempt (T14.91)']
data.sort_values('value', axis='index', ascending=False, inplace=True, na_position='last')
data = data.head(20)

data['angle'] = data['value']/data['value'].sum() * 2*pi
data['color'] = Category20c[20]

p = figure(plot_height=800, plot_width=1300, 
           title="Top 20 Primary Impressions for Opioids", 
          #  toolbar_location=None,
           tools="hover", 
           tooltips="@names: @value", 
           x_range=(-0.5, 1.0))

p.wedge(x=0, y=1, radius=0.4,
        start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        line_color="white", fill_color='color', legend="names", source=data)

p.axis.axis_label=None
p.axis.visible=False
p.grid.grid_line_color = None

show(p)
```

### Bar Plot of the counts of the top 20 Opioid related primary impressions

```
from bokeh.io import show, output_file
from bokeh.models import ColumnDataSource
from bokeh.palettes import Category20c
from bokeh.plotting import figure
from bokeh.transform import factor_cmap

# output_file("bar_colormapped.html")
source = ColumnDataSource(data)

p = figure(x_range=data.names, plot_height=800, plot_width=1000, 
           title="Top 20 Primary Impressions for Opioids",
           tooltips="@names: @value")
p.vbar(x='names', top='value', width=0.9, source=source,
       line_color='white', 
       fill_color=factor_cmap('names', palette=Category20c[20], factors=data.names))

# p.xgrid.grid_line_color = None
p.xaxis.major_label_orientation = pi/2.5


show(p)
```

## Exploring Data related to Suicide Attempts (T14.91)

```
suicides = v3_data[v3_data['Primary Impression ICD 10'] == 'T14.91']
suicides = suicides.rename(columns={'Year ': 'Year'})

print(suicides.columns)
```

```
sum_by_date = suicides.groupby(['Event Date'])['Count of Events'].sum()
suicides_df = sum_by_date.reset_index()
suicides_df.rename(columns={'Event Date': 'date', 'Count of Events': 'count'}, inplace=True)

suicides_df['date'] = pd.to_datetime(suicides_df['date'])
suicides_df.dtypes
```

```
import numpy as np

from bokeh.io import show
from bokeh.layouts import column
from bokeh.models import ColumnDataSource, RangeTool
from bokeh.plotting import figure

# dates = np.array(AAPL['date'], dtype=np.datetime64)
# source = ColumnDataSource(data=dict(date=dates, close=AAPL['adj_close']))
source = ColumnDataSource(data=suicides_df)


p = figure(plot_height=300, plot_width=800, tools="xpan", 
           toolbar_location=None,
           x_axis_type="datetime", x_axis_location="above",
           background_fill_color="#efefef",
           x_range=(suicides_df['date'][400], suicides_df['date'][500]),
           title="Counts of Suicide Attempts (Nemsis V3 data)")

p.line('date', 'count', source=source)
p.yaxis.axis_label = 'Count of Events'

select = figure(title="Drag the middle and edges of the selection box to change the range above",
                plot_height=130, plot_width=800, y_range=p.y_range,
                x_axis_type="datetime", y_axis_type=None,
                tools="", toolbar_location=None, background_fill_color="#efefef")

range_tool = RangeTool(x_range=p.x_range)
range_tool.overlay.fill_color = "navy"
range_tool.overlay.fill_alpha = 0.2

select.line('date', 'count', source=source)
select.ygrid.grid_line_color = None
select.add_tools(range_tool)
select.toolbar.active_multi = range_tool

show(column(p, select))
```
